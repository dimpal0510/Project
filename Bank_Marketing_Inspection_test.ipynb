# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types
# Check data types
print(df.dtypes)

# Check missing values
print(df.isnull().sum())

# Combined summary
print(df.info())

```

```python

# Summary statistics
print(df.describe())

```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
# Histogram + KDE curve
sns.histplot(df['y'], kde=True, bins=50)
plt.title("Distribution of Target Variable")
plt.xlabel("Target")
plt.ylabel("Count")
plt.show()
#sns.countplot(x='y', data=df)
#plt.title("Target Variable Distribution")
#plt.xlabel("Target (y)")
#plt.ylabel("Count")
#plt.show()
# sns.histplot(df['y'], kde=False)  # categorical, so KDE not useful
# plt.title("Distribution of Target Variable")
# plt.xlabel("Target (y)")
# plt.ylabel("Count")
# plt.show()
```

```python

# Correlation matrix for numerical features
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example: assume df is your DataFrame
# Select only numerical features
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Display correlation matrix
print(corr_matrix)

# Plot heatmap for better visualization
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix for Numerical Features")
plt.show()

```
